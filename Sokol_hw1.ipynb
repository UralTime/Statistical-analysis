# Задача 5

Функция распределения Вейбулла $ {\displaystyle F(x)=1-e^{-(x / \lambda)^a}}$ строго возрастает при $a>0, \lambda>0$, а её обратная функция имеет вид $ {\displaystyle F^{-1}(x)=\lambda \left(- \ln (1-x) \right)^{\frac{1}{a}}}. $

Таким образом, если ${\displaystyle U_{1},\ldots ,U_{n}}$ — выборка из стандартного непрерывного равномерного распределения, то $ X_{1},\ldots ,X_{n},$ где

${\displaystyle X_{i}=\lambda \left( - \ln (1-U_{i}) \right)^{\frac{1}{a}},\;i=1,\ldots ,n}$
— искомая выборка из распределения Вейбулла.


```python
import math
import matplotlib.pyplot as plt
import random
random.seed(42)



def my_weib(lmbda=1.0, a=5.0):
    return lmbda*(-math.log(1 - random.random()))**(1/a)

plt.hist([my_weib() for i in range(10000)], 50, density=True)
plt.title('Samples from hand rolled Weibull function')
```
