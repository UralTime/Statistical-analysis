# Задача 1

Пусть $\left\{u_t\right\}, t \in \mathbb{Z}$ - последовательность случайных величин, определенных на некотором вероятностном пространстве ( $\Omega, \mathcal{F}, P).$

Последовательность $\left\{u_t\right\}$ стационарна в узком смысле, если $\forall k \in \mathbb{N}, \forall \tau  \in \mathbb{Z}, \forall\left(t_1, \ldots, t_k\right) \in \mathbb{Z ^ k} $ :

$$
\operatorname{Law}\left(u_{t_1}, \ldots, u_{t_k}\right)=\operatorname{Law}\left(u_{t_1+\tau}, \ldots, u_{t_k+\tau}\right),
$$

где $\operatorname{Law}$ --- закон распределения вектора из случайных величин.

Последовательность $\left\{u_t\right\}$ стационарна в широком смысле, если $\forall t \in \mathbb{Z}, \forall \tau  \in \mathbb{Z}$ :

$$
\mathbb{E} u_t=\text { const } \quad \text{ и } \quad \operatorname{cov}\left(u_t, u_{t+\tau}\right)=f(\tau),
$$

где $f$ --- какая-то функция, зависящая только от $\tau.$




Последовательность $\left\{u_t\right\}$ - белый шум в узком смысле, если $\left\{u_t\right\}$ - набор независимых одинаково распределенных стандартных нормальных величин (с нулевым средним и единичной дисперсией)

Последовательность $\left\{u_t\right\}$ - белый шум в широком смысле, если $\forall t \in \mathbb{Z}$ $$\mathbb{E} u_t=0 \quad \text{ и } \quad \operatorname{cov}\left(u_t, u_s\right)=0  \text{ при }  t \ne s, \quad \operatorname{cov}\left(u_t, u_t\right)=1.$$

# Задача 3

- Акаике \begin{aligned}
&A I C=2 k-2 l\\
&\text { где } l \text { - значение логарифмической функции правдоподобия построенной модели, } k \text { - количество использованных (оцененных) параметров. }
\end{aligned}

- Шварца \begin{aligned}
&B I C=S C=k \ln n-2 l\\
&\text { Как видно из формулы, данный критерий налагает больший штраф на увеличение количества параметров по сравнению с AIC, так как } \ln n \text { больше } 2 \text { уже при количестве } 8 \text { наблюдений }
\end{aligned}

Известно, что для авторегрессионных моделей критерий AIC переоценивает порядок модели, то есть оценка порядка модели на основе этого критерия несостоятельна. Состоятельным критерием выбора порядка авторегрессионной модели является BIC.

Главным критерием того, что построенная линейная авторегрессионная модель хорошо подходит для рассматриваемого ряда, это факт, что остатки представляют из себя белый шум. Для этого используются соответствующие инструменты (тест Шапиро-Уилка, Q-Q plot).

Также можно сравнить информационные критерии AIC/BIC с другими моделями (чем меньше, тем лучше).

Пусть состояния процесса соответсвуют номерам строк в матрице. Тогда граф состояний выглядит так:

image.png

Стационарное распределение $\pi=\left(\pi_1, \pi_2, \ldots, \pi_N\right)$ существует, если:
1. Цепь неразложима (irreducible)
- Из любого состояния можно достичь любого другого (нет изолированных подмножеств).
2. Все состояния положительно возвратны (positive recurrent)
- Среднее время возврата в любое состояние конечно.
3. Матрица $Q$ удовлетворяет условию баланса
- $\pi Q=0$ (вектор $\pi$ - левый собственный вектор $Q$ с нулевым собственным числом).

Если граф сильно связный → цепь неразложима. В Python можно использовать networkx для проверки сильной связности графа.

```python
import numpy as np
import networkx as nx

Q = np.array([
    [-4, 0, 1, 3, 0],
    [0, -6, 0, 2, 4],
    [1, 0, -3, 0, 2],
    [1, 0, 0, -1, 0],
    [0, 1, 0, 0, -1]
])

G = nx.DiGraph(Q > 0)  # Строим граф переходов
print("Сильно связный?", nx.is_strongly_connected(G))  # True → неразложима
```


Для однородных процессов с конечным числом состояний: если цепь неразложима, то стационарное распределение существует и единственно.

В стационарном состоянии по уравнению Колмогорова $$0 = \frac{\partial p(t)}{\partial t} = p(t) Q,$$ где $Q$ --- матрица интенсивностей.



Находим левый собственный вектор $Q$ с нулевым собственным числом и нормируем, чтобы получился вектор вероятностей. Это и будет стационарное состояние

Ответ $\quad \left(\begin{array}{ccc}
\frac{3}{22} & \frac{1}{22} & \frac{1}{22} & \frac{1}{2} & \frac{3}{11}
\end{array}\right)$

#Задача 12

```python
import numpy as np
import matplotlib.pyplot as plt

# Матрица интенсивностей
Q = np.array([
    [-6, 3, 3, 0],
    [0, -2, 0, 2],
    [2, 0, -2, 0],
    [0, 2, 1, -3]
])

state = 0
n_steps = 50
trajectory = [state]

for _ in range(n_steps):
    rates = Q[state, :]
    # убрать отрицательные элементы
    rates = np.where(rates < 0, 0, rates)
    total_rate = np.sum(rates)

    # если нет возможных переходов
    if total_rate == 0:
        break

    time_to_next = np.random.exponential(1 / total_rate)

    # Выбор следующего состояния
    next_state = np.random.choice(len(rates), p=rates / total_rate)
    state = next_state
    trajectory.append(state)

# Визуализация траектории
plt.figure(figsize=(10, 6))
plt.plot(trajectory, marker='o')
plt.xlabel('Шаг')
plt.ylabel('Состояние')
plt.yticks([0, 1, 2, 3], ['1 сервер', '2 сервер', '3 сервер', '4 сервер'])
plt.grid()
```
